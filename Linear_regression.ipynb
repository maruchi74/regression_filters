```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset=np.loadtxt('ge3_data_all_beta_nolabel.csv', delimiter=',')
x_raw=dataset[:,0:9]
y_raw=dataset[:,9]
y_raw=np.reshape(y_raw, (-1,1))
n_row = x_raw.shape[0]

x=x_raw
y=y_raw
```

```python
model = LinearRegression().fit(x, y)
```

```python
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq),
print('intercept:', model.intercept_),
print('slope:', model.coef_)
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(18,4))
plt.plot(y_pred, label='predicted')
plt.plot(y, label='true')
plt.legend();
plt.grid()
```

```python
plt.figure(figsize=(18,4))
plt.plot(y_pred[800:1200,0], label='predicted')
plt.plot(y[800:1200,0], label='true')
plt.legend();
plt.grid()
```

```python
rmse = np.sqrt(mean_squared_error(y,y_pred))
r2 = r2_score(y,y_pred)
print(rmse)
print(r2)
```

Linear Regression by Polynomial

```python
transformer = PolynomialFeatures(degree=3, include_bias=False)
```

```python
transformer.fit(x)
```

```python
x_ = transformer.transform(x)
```

```python
x_ = PolynomialFeatures(degree=3, include_bias=False).fit_transform(x)
```

```python
model = LinearRegression().fit(x_, y)
```

```python
r_sq = model.score(x_, y)
print('coefficient of determination:', r_sq),
print('intercept:', model.intercept_),
print('slope:', model.coef_)
```

```python
y_pred_ = model.predict(x_)
print('predicted response:', y_pred, sep='\n')
```

```python

plt.figure(figsize=(18,4))
plt.plot(y_pred_, label='predicted')
plt.plot(y, label='true')
plt.legend();
plt.grid()
```

```python
plt.figure(figsize=(18,4))
plt.plot(y_pred_[800:1200,0], label='predicted')
plt.plot(y[800:1200,0], label='true')
plt.legend();
plt.grid()
```

```python
plt.figure(figsize=(18,4))
plt.plot(y_pred_[500:1000,0], label='predicted')
plt.plot(y[500:1000,0], label='true')
plt.legend();
plt.grid()
```

```python
rmse_ = np.sqrt(mean_squared_error(y,y_pred_))
r2_ = r2_score(y,y_pred_)
print(rmse_)
print(r2_)
```
